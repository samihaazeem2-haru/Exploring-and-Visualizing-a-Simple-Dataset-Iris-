```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset from seaborn
df = sns.load_dataset("iris")

# Display first few rows
df.head()

# Save the dataset
# df.to_csv("iris.csv", index=False)
```

Basic information

```python
print("Shape of dataset:", df.shape)
print("\nColumn Names:\n", df.columns)

print("\nDataset Info:\n")
df.info()

print("\nSummary Statistics:\n")
df.describe()
```

Data Visualization

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x="sepal_length",
    y="petal_length",
    hue="species",
)
plt.title("Sepal Length vs Petal Length")
plt.show()

# Sepal and petal measurements form clear clusters for each species.
```

Histograms (Value Distributions)

```python
df.hist(figsize=(10, 8), color="skyblue")
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()

# Histograms show normal-like distributions for most features
```

Box Plots (Outlier Detection)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Box Plot of All Features")
plt.show()

# Boxplots reveal very few outliers, confirming the dataset is clean and balanced.
```
